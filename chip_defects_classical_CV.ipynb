Fourier

```python
from scipy import fftpack

# Apply Fourier Transform
f_transform = fftpack.fft2(image_sym)

# Shift zero frequency components to the center
f_transform_shifted = fftpack.fftshift(f_transform)

# Calculate magnitude spectrum
magnitude_spectrum_sym = np.abs(f_transform_shifted)
```

```python

# Apply Fourier Transform
f_transform = fftpack.fft2(image_zalip)

# Shift zero frequency components to the center
f_transform_shifted = fftpack.fftshift(f_transform)

# Calculate magnitude spectrum
magnitude_spectrum_zalip = np.abs(f_transform_shifted)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].imshow(np.log(1 + magnitude_spectrum_sym), cmap='gray')
ax[0].set_title('Симметрия')

ax[1].imshow(np.log(1 + magnitude_spectrum_zalip), cmap='gray')
ax[1].set_title('Залипание')

plt.show()
```

Hough Transform

```python

# Apply edge detection (you may want to use other edge detection methods based on your specific case)
edges = cv2.Canny(image_sym, 10, 150, apertureSize=3)

# Use Hough Line Transform to detect lines in the image
lines = cv2.HoughLines(edges, 1, np.pi / 180, threshold=250)

# Plot the original image and the detected lines
plt.subplot(1, 2, 1)
plt.imshow(image_sym, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.axis([0, 640, 0, 480])

# Plot the detected lines
for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    plt.plot([x1, x2], [y1, y2], 'r')

plt.title('Hough Line Transform')
plt.show()
```
